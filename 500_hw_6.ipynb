```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def baro_ins(B, U, L):
    y = np.linspace(-L, L, 25)
    part_1 = (U*np.pi**2)/(2*L**2)
    part_sin = (np.sin((np.pi*y)/(2*L))**2)
    part_cos = (np.cos((np.pi*y)/(2*L))**2)
    return(B-(part_1*(part_sin-part_cos)))
```

```python
condition_low_lat = baro_ins(B=2.254304E-11, U=10.0, L=10.0E6) # 10 deg N
condition_high_lat = baro_ins(B=7.829115E-12, U=10.0, L=10.0E6) # 70 deg N
# using 45 degrees north for rest of problem
condition_weak_jet = baro_ins(B=1.618624E-11, U=1.0, L=10.0E6) 
condition_strong_jet = baro_ins(B=1.618624E-11, U=100.0, L=10.0E6)
condition_thin_jet = baro_ins(B=1.618624E-11, U=10.0, L=5.0E6)
condition_wide_jet = baro_ins(B=1.618624E-11, U=10.0, L=15.0E6)
```

```python
fig = plt.figure(figsize=(20,20))
ax1 = plt.subplot(3, 1, 1)
ax2 = plt.subplot(3, 1, 2)
ax3 = plt.subplot(3, 1, 3)
ax1.plot(np.linspace(-10.0E6, 10.0E6, 25), condition_low_lat, color='blue', label='Low Latitude', linewidth=5)
ax1.plot(np.linspace(-10.0E6, 10.0E6, 25), condition_high_lat, color='orange', label='High Latitude', linewidth=5)
ax2.plot(np.linspace(-10.0E6, 10.0E6, 25), condition_weak_jet, color='green', label='Weak Jet', linewidth=5)
ax2.plot(np.linspace(-10.0E6, 10.0E6, 25), condition_strong_jet, color='red', label='Strong Jet', linewidth=5)
ax3.plot(np.linspace(-5.0E6, 5.0E6, 25), condition_thin_jet, color='purple', label='Thin Jet', linewidth=5)
ax3.plot(np.linspace(-15.0E6, 15.0E6, 25), condition_wide_jet, color='yellow', label='Wide Jet', linewidth=5)
ax1.legend()
ax2.legend()
ax3.legend()
```
