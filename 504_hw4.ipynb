# 1)
The unit mass extinction coefficient is a quantity of area over mass that describes the contribution of extinction by the reference particle. The unit volume extinction coefficient is a quantity of one over meters, which is the product of number of particles per unit volume air, areal cross section of each particle, and scattering/absorption efficiency. It can also be defined as the product of air density, mass of absorber/scatterer per unit volume, and the mass extinction coefficient. They are related through the fact that the density of the absorber multiplied by the mass extinction coefficient is equal to the volume extinction coefficient. 
Since:
$$
dI_{\lambda} =  \rho \cdot \sigma_{a\lambda} \cdot I_{\lambda} \cdot ds
$$
And:
$$
dI_{\lambda} =  \rho \cdot r \cdot k_{\lambda} \cdot I_{\lambda} \cdot ds
$$
And V, the volume absorption coefficient, can be defined as:
$$
V = \rho \cdot r \cdot k_{\lambda}
$$
Then, we can relate the mass absorption coefficient:
$$ 
\sigma_{a\lambda} 
$$
To the volume absorption coefficient:
$$
V = \rho \cdot \sigma_{a\lambda}
$$

(Wallace and Hobbs 4.4)

The atmospheric normal optical depth is a measure of the cumulative depletion that a beam of radiation directed downward would experience in passing through the layer. (Wallace and Hobbs 4.5)

# 2b)
Optical Depth:
$$
\tau_\lambda = 
ln(F_{TOA}) - ln(F_{BOA})
$$ 
Transmissivity:
$$
T_\lambda = e^{-\tau} = e^{-(ln(F_{TOA}) - ln(F_{BOA}))}
$$

# 2c)
$$
T = e^{-\tau/sin(\theta)}
$$
if the elevation angle is: $$ \theta $$ relative to the surface horizon, then the solar zenith angle is: $$ 90 - \theta $$
given that the formula for transmissivity with light incoming at an angle relative to surface normal is: 
$$
T = e^{-\tau/cos(\theta)}
$$
laws of trigonometry tell us that:
$$ sin(\theta) = cos(90-\theta) $$
thus, we show that:
$$
T = e^{-\tau/sin(\theta)}
$$
