## 4. Problem finish the code by adding an if-else block into the code to subsample the years and temperature data (temp1) that have temperature anomalies > 0. Then add another barplot to show these temperatures in the plot as bars  with a different color. (20pts)

    

The first four lines of code set four variables to empty sets. The next line assigns the variable n to the length of the set found under the variable temp1. Next, i is set to 0, and the while loop begins after that. It assigns 2 variables at the index "i" in each array, and then I wrote an if else block. The "if" part test for if t, the temperature anomaly, is less than 0. If it is, the numpy package appends the array in t to include only those that are negative. The next line attatches the corresponding year. The "elif" part test for when t is greater than 0, and edits the array so that the positive values are in it, and attatches the corresponding year. The print statements print which anomalies were found. 

The next section of code, related to the graph, starts by changing the size of the plot so it is slightly longer. Line 3 uses matplotlib to create a bar graph of the two data arrays, and chooses a color and a label. The next few lines of code create labels and ticks on the edge of the graph, as well as set axis limits. The plt.grid line creates a gridspace in the background of the graph. Lines 14 through 16 are essentialy a repeat of the previous lines, but for the positive anomalies. 

```python
year_neg= [ ]
temp1_neg = [ ]
year_pos = [ ]
temp1_pos = [ ]
n=len(temp1)
i=0
while i<n:
    t=temp1[i]
    y=year[i]
    if t<0:
        temp1_neg.append(t) # add the negative temperature anomaly to the list
        year_neg.append(y)  # add the corresponding year to the list
    elif t>0:
        temp1_pos.append(t) # add the positive temperature anomaly to the list
        year_pos.append(y)  # add the corresponding year to the list    
    i=i+1 # increase our index counter
   
# end of for loop
print("found "+ str(len(temp1_neg))+" negative amomalies in the GMST time series")
print("found "+ str(len(temp1_pos))+" positive amomalies in the GMST time series")
```

```python
# cell with cod for plotting the time series as a barplot
plt.figure(figsize=[12,6])
plt.bar(year_neg,temp1_neg,color='mediumslateblue',label='negative anomalies')
plt.legend(fontsize=18)
plt.xlabel("year",fontsize=18)
plt.ylabel("anomalies [deg C]",fontsize=18)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.ylim([-0.5,0.8])
plt.xlim([1880,2020])
plt.title("NASA GISS GMST",fontsize=18)
plt.grid()

plt.bar(year_pos,temp1_pos,color="maroon",label="positive anomalies")
plt.legend(fontsize=18)
plt.show()
```
