```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

*1)*

```python
Lambda=np.arange(.02E-6,4.02E-6,.02E-6)

C1=3.74151E-22
C2=1.43879E-2
T=6000

y=[]
for i in range(0,200):
    x=C2/(Lambda[i]*T)
    I=C1/((3.1415926*Lambda[i]**5)*((math.exp(x))-1))
    y.append(I)

for i in range(0,200):
    print("%.8f" % Lambda[i],"%.2f" % y[i])
    
plt.plot(Lambda,y)
plt.xlabel('Wavelength (m)')
plt.ylabel('Spectral Emitted Radiance (T=6000K)')
plt.title('Question 1')
```

*2)*

```python
Lambda1=np.arange(.5E-6,100.0E-6,.5E-6)

C1=3.74151E-22
C2=1.43879E-2
T1=300

z=[]
for i in range(0,200):
    x1=C2/(Lambda1[i]*T1)
    I1=C1/((3.1415926*Lambda1[i]**5)*((math.exp(x1))-1))
    z.append(I1)

for i in range(0,200):
    print("%.8f" % Lambda1[i],"%.2f" % z[i])
    
plt.plot(Lambda1,z)
plt.xlabel('Wavelength (m)')
plt.ylabel('Spectral Emitted Radiance (T=300K)')
plt.title('Question 2')
```

*3)*

```python
h=6.62606896E-34
c=2.9979E8

def photon(Lambda,y):
    return(y/((h*c)/Lambda))

i=0
g=0
p=[]
while np.logical_and(i<200,g<200):
    p.append(photon(Lambda[i],y[g]))
    i=i+1
    g=g+1
    
plt.plot(Lambda,p)
plt.xlabel('Wavelength (m)')
plt.ylabel('Number of Photons')
plt.title('Photons (T=6000K)')
plt.show()

n=0
k=[]
d=0
while np.logical_and(n<200,d<200):
    k.append(photon(Lambda1[n],z[d]))
    n=n+1
    d=d+1
    
    
plt.plot(Lambda1,k)
plt.xlabel('Wavelength (m)')
plt.ylabel('Number of Photons')
plt.title('Photons (T=300K)')
plt.show()
```

*4)*

```python
print(' Plotted Wavelength of Maximum Radiance (T=6000K): ' + str(Lambda[np.argmax(y)]) + ' m ')
print(' Plotted Wavelength of Maximum Radiance (T=300K): ' + str(Lambda1[np.argmax(z)]) + ' m ')

# Wien's Law Calculations

def Wien(T):
    return(2897.79E-6/T)

print(' Calculated Wavelength of Maximum Radiance (T=6000): ' + str(Wien(6000)) + ' m ')
print(' Calculated Wavelength of Maximum Radiance (T=300):' + str(Wien(300)) + ' m ')
```

The maximum wavelengths from the plot are the same as the maximum wavelengths calculated from Wien's Law.
