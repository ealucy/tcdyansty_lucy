```python
real_time = pd.to_datetime(df['Time (hhmmss)'], format='%H%M%S')
```

```python
real_time
```

```python
plt.plot(real_time, wspd)
```

```python
real_time = real_time[750:851]
```

```python
fig = plt.figure(figsize=(10,8))
ax1 = plt.subplot(2,2,1)
ax1.plot(real_time, w, color='teal')
ax1.set_ylabel('Vertical Velocity (m/s)')
ax2 = plt.subplot(2,2,2)
ax2.plot(real_time, wpert, color='dodgerblue')
ax2.set_ylabel('Vertical Velocity Perturbations (m/s)')
ax3 = plt.subplot(2,2,3)
ax3.plot(real_time, wspd_sel, color='mediumvioletred')
ax3.set_ylabel('Wind Speed (m/s)')
ax4 = plt.subplot(2,2,4)
ax4.plot(real_time, psfc_pert, color='goldenrod')
ax4.set_ylabel('Surface Pressure Perturbations (hPa)')

ax1.grid(True)
ax1.xaxis_date()
ax1.tick_params(axis='x', rotation=45, labelsize=7.5)

ax2.grid(True)
ax2.xaxis_date()
ax2.tick_params(axis='x', rotation=45, labelsize=7.5)

ax3.grid(True)
ax3.xaxis_date()
ax3.tick_params(axis='x', rotation=45, labelsize=7.5)

ax4.grid(True)
ax4.xaxis_date()
ax4.tick_params(axis='x', rotation=45, labelsize=7.5)
```

```python
theta = df['Theta (K)'][750:851]
theta_e = df['Theta_E (K)'][750:851]
```

```python
fig = plt.figure(figsize=(10,4))
ax1 = plt.subplot(1,2,1)
ax1.plot(real_time, theta_e, color='slateblue')
ax1.set_ylabel('Equivalent Potential Temperature (K)')
ax1.grid(True)
ax1.xaxis_date()
ax1.tick_params(axis='x', rotation=45, labelsize=7.5)
```

```python
fig, ax1 = plt.subplots()

color = 'slateblue'
ax1.set_ylabel('Equivalent Potential Temperature (K)')
ax1.plot(real_time, theta_e, color=color, label='theta_e')
ax1.tick_params(axis='y')
ax1.tick_params(axis='x', rotation=45)
ax1.legend(loc='upper left', fontsize=8)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'goldenrod'
ax2.set_ylabel('Surface Pressure Perturbations (hPa)')  # we already handled the x-label with ax1
ax2.plot(real_time, psfc_pert, color=color, label='p_pert')
ax2.tick_params(axis='y')
ax2.legend(loc='upper right', fontsize=8)

fig.tight_layout()# otherwise the right y-label is slightly clipped
plt.show()
```

```python
wdir = df['Direction (°)'][750:851]
wspd = wspd[750:851]
```

```python
fig, ax1 = plt.subplots(figsize=(10,5))

color = 'red'
ax1.set_ylabel('Wind Speed (m/s)')
ax1.plot(real_time, wspd, color=color)
ax1.tick_params(axis='y')
ax1.tick_params(axis='x', rotation=45)
ax1.grid(True)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'black'
ax2.set_ylabel('Wind Direction (deg)')  # we already handled the x-label with ax1
ax2.scatter(real_time, wdir, color=color, s=8)
ax2.tick_params(axis='y')


fig.tight_layout()# otherwise the right y-label is slightly clipped
plt.show()
```
