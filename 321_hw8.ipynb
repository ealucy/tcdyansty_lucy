```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
def R(RHcs):
    return((2*.075)/(np.log(RHcs)*1000*461*300))
```

```python
RHcs=[101,111,300]
i=0
radii=[]
for x in RHcs:
    radii.append(R(RHcs[i]))
    i=i+1
```

```python
T = 273
m = 58.5
M = 1.0e-19
i = 2
mw = 18

a = (2*.075)/(1000*461*T)
b = (3*i*mw*M)/(4*np.pi*1000*m)
```

```python
def curv(R):
    return(1+(a/R))
def solute(R):
    return(1-(b/R**3))
def kohler(R):
    return(1+((a/R)-(b/R**3)))
```

```python
x = np.arange(.025e-6 , 2.0025e-6 , .0025e-6)
y = np.arange(.025e-6 , 2.0025e-6 , .0025e-6)
```

```python
h=0
curvatureplot=[]
for value in x:
    curvatureplot.append(curv(x[h]))
    h=h+1
g=0
solutionplot=[]
for value in x:
    solutionplot.append(solute(x[g]))
    g=g+1
f=0
kohlerplot=[]
for value in y:
    kohlerplot.append(kohler(y[f]))
    f=f+1
```

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(1,1,1)
ax.plot(x , curvatureplot , color = 'red' , label = 'Curvature Effect')
ax.plot(x , solutionplot , color = 'blue' , label = 'Solute Effect')
ax.plot(y , kohlerplot , color = 'yellow' , label = 'Kohler Curve')
ax.legend()
ax.set_xlabel('Radius (micrometers)')
ax.set_ylabel('RHhs (%)')
ax.set_ylim(.98,1.02)
ax.set_title('The dependence of equilibrium saturation ratio RHcs as a function of the radius R of a solution droplet containing fully dissociated NaCl')
```

```python
a1 = (2*.075)/(1000*461*280)
b1 = (3*2*18*1e-19)/(4*np.pi*1000*58.5)

a2 = (2*.075)/(1000*461*280)
b2 = (3*1*18*1e-19)/(4*np.pi*1000*17)

RmaxNACL = (3*b1/a1)**(1/2)
RmaxNH3 = (3*b2/a2)**(1/2)

RHhsmaxNACL = (1+(a1/RmaxNACL)-(b1/RmaxNACL**3))
RHhsmaxNH3 = (1+(a2/RmaxNH3)-(b2/RmaxNH3**3))
```

```python
print('the Rmax for NACL is' , RmaxNACL , 'and its RHhsmax is' , RHhsmaxNACL)
print('the Rmax for NH3 is' , RmaxNH3 , 'and its RHhsmax is' , RHhsmaxNH3)
print('NH3 can be nucleated more easily because it has a smaller RHhsmax. ')
print('No matter what the enviornmental RH, as long as RHhs is smaller, the particle will nucleate.')
print('So, whichever of these two has a smaller RHhsmax will nucleate more easily. This is NH3')
```

```python
rads = np.arange(.025 , 2.05 , .05)
S = .05
a = (2*.075)/(1000*461*273)
b = (3*2*18*1e-18)/(4*np.pi*1000*58.5)
Fk = 7.643e10
Fd = 9.761e6
def drdt(radius):
    return((S - (a/radius) + (b/(radius)**3))/(radius*(Fk + Fd)*1000))
d = 0 
nums = []
for value in rads:
    nums.append(drdt(rads[d]))
    d = d + 1
print(nums)
```

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(1,1,1)
ax.plot(rads , nums)
ax.set_xlabel('Radius (micrometers)')
ax.set_ylabel('Growth Rate (micrometers/second)')
ax.set_title('Change in growth rate as radius changes')
```

```python
print('As radius increases, condensational growth rate decreases exponentially.')
print('The growth rate is fast for when radius is very small.')
print('It becomes very slow when radius is large.')
```
