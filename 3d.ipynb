```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x, y):
    return (x*y)

x = np.arange(0, 10, 2)
y = np.arange(0, 10, 2)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

```python
ax.view_init(15, 5) # sets elevation and azimuthal angle viewing points
fig
```
