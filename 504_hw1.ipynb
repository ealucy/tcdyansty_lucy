##  <span style="color:darkblue">C. Calculations</span> 

```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

# Set up variables for surface temperature.

```python
t_earth = 300
t_sun = 6000
```

# <span style="color:darkblue">1)</span> 

# Set up array of wavelengths and constants, set up numerator and denominator in spectral emitted radiance equation, calculate spectral emitted radiance.

# The Planck Function


$$
B_\lambda = ({c_1}\cdot{\lambda^-5})\frac{1}{\pi(e^{\frac{c_2}{\lambda \cdot T}} - 1)}
$$




```python
wavelength_range_1 = np.arange(.02E-6, 4.02E-6, .02E-6)
c1 = 3.74E-16
c2 = 1.45E-2
num = c1*(wavelength_range_1**-5)
denom = np.pi*(math.e**(c2/(t_sun*wavelength_range_1))-1)
spectral_emitted_radiance_1 = num/denom
```

# Plot the computed spectral emitted radiance as a function of wavelength.

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(1,1,1)
ax.plot(wavelength_range_1, spectral_emitted_radiance_1, color='red', linewidth=3, label='T=6000K')
ax.set_xlabel('wavelength (m)')
ax.set_ylabel('spectral emitted radiance (W/m^2/sr)')
ax.legend();
```

# <span style="color:darkblue">2)</span> 

# Set up array of wavelengths and constants, set up numerator and denominator in spectral emitted radiance equation, calculate spectral emitted radiance.

# The Planck Function


$$
B_\lambda = ({c_1}\cdot{\lambda^-5})\frac{1}{\pi(e^{\frac{c_2}{\lambda \cdot T}} - 1)}

```python
wavelength_range_2 = np.arange(.5E-6, 100.5E-6, .5E-6)
c1 = 3.74E-16
c2 = 1.45E-2
num = c1*(wavelength_range_2**-5)
denom = np.pi*(math.e**(c2/(t_earth*wavelength_range_2))-1)
spectral_emitted_radiance_2 = num/denom
```

# Plot the computed spectral emitted radiance as a function of wavelength.

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(1,1,1)
ax.plot(wavelength_range_2, spectral_emitted_radiance_2, color='red', linewidth=3, label='T=300K')
ax.set_xlabel('wavelength (m)')
ax.set_ylabel('spectral emitted radiance (W/m^2/sr)')
ax.legend();
```

# <span style="color:darkblue">3)</span> 

# Find the array location of wavelength of maximum radiance and use that to find the wavelength of maximum radiance from the calculation. Use Wien's Law to calculate the same variable, then compare for each 1 and 2.

# Wiens Law

$$
\lambda_{\text{max}} = \frac{2897}{T}
$$

```python
I_max_loc_1 = np.where(spectral_emitted_radiance_1==spectral_emitted_radiance_1.max())[0][0]
wavelength_max_1_direct = wavelength_range_1[I_max_loc_1]
wavelength_max_1_calc = 2897E-6/t_sun
I_max_loc_2 = np.where(spectral_emitted_radiance_2==spectral_emitted_radiance_2.max())[0][0]
wavelength_max_2_direct = wavelength_range_2[I_max_loc_2]
wavelength_max_2_calc = 2897E-6/t_earth
```

```python
print('question 1 wavelength of maximum radiance:') 
print(str(wavelength_max_1_direct))
print('wien:') 
print(str(wavelength_max_1_calc))
```

```python
print('question 2 wavelength of maximum radiance:') 
print(str(wavelength_max_2_direct))
print('wien:') 
print(str(wavelength_max_2_calc))
```

the maximum wavelengths given by each method are very similar

the highest frequency band found was 52.6 GHz, or 5.26E10 Hz, from https://www.sdxcentral.com/5g/definitions/what-is-5g/what-is-5g-spectrum/

# Wavelength

$$
\lambda = {c}/{{f}}
$$

using c=3E8 m/s

```python
frequency = 5.26E10 
c = 3E8
wavelength = c/frequency
print('the wavelength of that frequency band is ' + str(wavelength) + 'm')
```
