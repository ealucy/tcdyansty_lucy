```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
T=np.arange(-80.0+273,2.0+273,2.0)
T
```

```python
def esw(T):
    return(6.11*math.exp(53.49-(6808/T)-5.09*math.log(T)))
esw(273.15)
```

```python
def esi(T):
    return(6.11*math.exp(26.16-(6293/T)-.555*math.log(T)))
```

```python
i=0
eswater=[]
esice=[]
for value in T:
    eswater.append(esw(T[i]))
    esice.append(esi(T[i]))
    i=i+1
```

```python
temps=np.arange(-80.0,2.0,2.0)
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1,1,1)
ax.plot(temps,eswater,color='#8f0327',label='esWater')
ax.plot(temps,esice,color='#01459e',label='esIce')
ax.legend(loc='best')
ax.set_xlabel('Temperature (Degrees Celsius)')
ax.set_ylabel('Vapor Pressure (mb)')
```

```python
eswater=np.array(eswater)
esice=np.array(esice)
difference=eswater-esice
print(np.amax(difference))
difference
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1,1,1)
ax.plot(temps,difference,color='#32007b')
ax.set_xlabel('Temperature (Degrees Celsius)')
ax.set_ylabel('Vapor Pressure (mb)')
ax.set_title('difference in vapor pressure as a function of temperature between eswater and esice')
ax.scatter(temps[34],0.27969684705224873,label='temperature at which difference maximises',color='k')
ax.legend(loc='upper left')
```

```python
print('the temperature at which the difference maximises is' , temps[34] , 'degrees celsius')
```
